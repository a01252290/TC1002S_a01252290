```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as np
```

```python
maps = pd.read_csv("Pace of life.csv", na_values = ":")
maps
```

```python
x1 = maps.Bank
y1 = maps.Talk
plt.scatter (x1,y1, color = "blue")
plt.x1label("Bank")
plt.y1label("Talk")
```

```python
sns.set(rc = {"figure.figsize":(11.7,8.27)})
sns.displot(maps["Bank"], bins = 35)
plt.show()
```

```python
mapscalor = maps.corr().round(2)
sns.heatmap(data = mapscalor, annot = True)
```
